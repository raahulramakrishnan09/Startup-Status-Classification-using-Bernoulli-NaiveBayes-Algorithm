```python
#filling_the_missing_values_using_fillna
df['Unnamed: 6']=df['Unnamed: 6'].fillna(df['Unnamed: 6'].mode()[0])
df['closed_at']=df['closed_at'].fillna(df['closed_at'].mode()[0])
df['age_first_milestone_year']=df['age_first_milestone_year'].fillna(df['age_first_milestone_year'].mean())
df['age_last_milestone_year']=df['age_last_milestone_year'].fillna(df['age_last_milestone_year'].mean())
```

```python
#dropping_rows
df.dropna(inplace=True)
```

```python
#label_encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
columns=df.select_dtypes(include=['object']).columns
for col in columns:
  df[col]=a.fit_transform(df[col])
```

```python
#heatmap
plt.figure(figsize=(20,15))
corr=df.corr()
sns.heatmap(corr,annot=True,fmt='.2g',linewidth=1)
plt.tight_layout()
plt.show()
```

```python
x=df.drop('status',axis=1)
y=df['status']
```

```python
#training_and_testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
x_train
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
f=StandardScaler()
x_train=f.fit_transform(x_train)
```

```python
#BernoulliNB_algorithm
from sklearn.naive_bayes import BernoulliNB
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
y_pred=bnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
#testing_accuracy_score
accuracy_score(y_pred,y_test)
```

```python
#confusion_matrix
confusion_matrix(y_pred,y_test)
```
